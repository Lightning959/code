```python
PATCH_SIZE = 60
textures = [(50,110),(250, 310),(450, 510),(650, 710),(850, 910)]
grass_patches = []
for loc in textures:
    grass_patches.append(gray[loc[0]:loc[0] + PATCH_SIZE,
                               loc[1]:loc[1] + PATCH_SIZE])
print(len(grass_patches[0]))
print(grass_patches[0].shape)
```
